```python
import numpy as np
import pandas as pd

# æ‰“å¼€csvæ–‡ä»¶
df = pd.read_csv("BeijingPM20100101_20151231.csv", encoding="gbk", dtype=str)

df
```

```python
# å»ºç«‹æ–°ç´¢å¼•
df['avg'] = 0
sum = 0
count = 0

# è®¡ç®—df['sum']å’Œdf['count']
for i in range(52582):
    sum = 0
    count = 0
    if not(df['PM_Dongsi'][i] is np.nan):
        sum += int(df['PM_Dongsi'][i])
        count += 1
    if not(df['PM_Dongsihuan'][i] is np.nan):
        sum += int(df['PM_Dongsihuan'][i])
        count += 1
    if not(df['PM_Nongzhanguan'][i] is np.nan):
        sum += int(df['PM_Nongzhanguan'][i])
        count += 1
    if not(df['PM_US Post'][i] is np.nan):
        sum += int(df['PM_US Post'][i])
        count += 1
    if(count == 0):
        count = 1
    # print(sum, count)
    # è®¡ç®—df['avg']
    df['avg'][i] = round(sum/count, 2)
    # print(df['avg'][i])
# df.to_csv("BeijingPM-avg.csv", encoding="gbk")
```
