```python
import pandas as pd
import requests
import json
import geocoder
```

```python
df = pd.read_csv('2.csv', encoding = 'gbk')
```

```python
df
```

```python
for i in df.index:
    
    baiduUrl = "http://api.map.baidu.com/geocoder/v2/?ak=vSpq0zwhjlzDsK8IG5ox7nZNpvs1chjd&callback=renderReverse&location={},{}&output=json&pois=0".format(df.loc[i, '经度'],df.loc[i, '纬度']) 
    req = requests.get(baiduUrl)
#     print(baiduUrl)
    content = req.text
    content = content.replace("renderReverse&&renderReverse(", "")
    content = content[:-1]
#     print(content)
    baiduAddr = json.loads(content)
    province = baiduAddr["result"]["addressComponent"]["province"]
    city = baiduAddr["result"]["addressComponent"]["city"]

    district = baiduAddr["result"]["addressComponent"]["district"]
    df.loc[i, 'province'] = province
    df.loc[i, 'city'] = city
```
