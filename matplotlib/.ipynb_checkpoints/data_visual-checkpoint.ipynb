```python
import numpy as np
import pandas as pd
%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
plt.style.use('seaborn-dark')
fig, ax = plt.subplots()
ax.set_title("cube numbers")
l = list(range(-10, 11))
x = np.array(l)
# array一个一维矩阵，相当于可以直接相乘
y = x * x * x
plt.bar(x, y)
for a, b in zip(x, y):
    plt.text(a, b * 0.5, '{}'.format(b), ha = 'center', va = 'bottom', fontsize = 10)
```
