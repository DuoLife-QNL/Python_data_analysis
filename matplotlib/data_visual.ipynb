### 创建数据并查看其分布情况（尝试创建了正态分布）

```python
# 创建一个满足正态分布的数据，均值为700，标准差为100，规模10000个
samples = np.random.normal(loc = 500, scale = 100, size = 10000)
# 定义函数，完成对某个list中的数值进行统计，并返回一个dict
def count_elements(sample):
    value_count = {}
    for x in sample:
        value_count[int(x)] = value_count.get(int(x), 0) + 1
    return value_count
# 绘制柱状图
counted = count_elements(samples)
fig1 = plt.figure()
ax1 = fig1.add_subplot(111)
ax1.axis([0, 1000, 0, 60])
ax1.bar(list(counted.keys(), counted.values())
```

# 作业6

导入数据

```python
Beijing = pd.read_csv('../data/BeijingPM20100101_20151231.csv')
Chengdu = pd.read_csv('../data/ChengduPM20100101_20151231.csv')
Guangzhou = pd.read_csv('../data/GuangzhouPM20100101_20151231.csv')
Shanghai = pd.read_csv('../data/ShanghaiPM20100101_20151231.csv')
Shenyang = pd.read_csv('../data/ShenyangPM20100101_20151231.csv')
haze = [Beijing, Guangzhou, Shenyang, Shanghai, Chengdu]
```

```python
# 数据预处理，丢弃空值数据，并计算平均值
# 由于几个城市的PM监测点数目不同，故需要传递描述PM指数的起始列和结束列
def data_pre(city, PM_col_start, PM_col_end):
    s_col = city.columns[1:4].append(city.columns[PM_col_start : PM_col_end + 1])
    PM_col = city.columns[PM_col_start : PM_col_end + 1]
    city = city.loc[:, s_col]
    city_cln = city.dropna(axis = 0, how = 'all', subset =PM_col).copy()
    city = city_cln
    city.loc[:, 'PM_AVG'] = np.nan
    for i in city.index:
        cnt_not_null = 0
        PM_sum = 0
        for j in city.columns[3 : PM_col_end - 2]:
            if not np.isnan(city.loc[i, j]):
                cnt_not_null += 1
                PM_sum += city.loc[i, j]
        city.loc[i, 'PM_AVG'] = PM_sum / cnt_not_null
    return city
```

```python
%%time
# 先处理北京（有四列数据和PM相关）
haze[0] = data_pre(haze[0], 6, 9)
# 再处理其它的城市（只有三列和PM有关）
for x in range(1, 5):
    haze[x] = data_pre(haze[x], 6, 8)
```

```python
# 按照天聚集
grouped = []
for x in range(0, 5):
    grouped.append(haze[x]['PM_AVG'].groupby([haze[x]['year'], haze[x]['month'], haze[x]['day']]).mean())
```

```python
df = []
for x in range(0, 5):
    df.append(pd.DataFrame(grouped[x]))
    for i in df[x].index:
        df[x].loc[i, 'date'] = str(i[0]) + '-' + str(i[1]) + '-' + str(i[2])
```

**将全部数据绘制到图上**

```python
fig6 = plt.figure()
bj = fig6.add_subplot(231)
cd = fig6.add_subplot(232)
gz = fig6.add_subplot(233)
sh = fig6.add_subplot(234)
sy = fig6.add_subplot(235)
bj.plot(df[0].date, df[0].PM_AVG)
cd.plot(df[1].date, df[1].PM_AVG)
gz.plot(df[2].date, df[2].PM_AVG)
sh.plot(df[3].date, df[3].PM_AVG)
sy.plot(df[4].date, df[4].PM_AVG)
bj.set_title('Beijing')
cd.set_title('Chengdu')
gz.set_title('Guangzhou')
sh.set_title('Shanghai')
sy.set_title('Shenyang')
```

**但像上面这样将六年的数据全部展示在一张图上导致可视化效果极差，故下面以2014年7月各城市的PM变化曲线为例进行展示**

```python
city = []
for x in df:
    result = pd.DataFrame()
    for y in range(1, 32):
        if (2014, 7, y) in x.index:
            result.loc[y, 'PM'] = x.loc[(2014, 7, y), 'PM_AVG']
        else:
            result.loc[y, 'PM'] = np.nan
    city.append(result)
```

```python
plt.style.use('default')
fig7 = plt.figure()
ax = []
for x in range(0, len(city)):
    ax.append(fig7.add_subplot(2, 3, x + 1))
    ax[x].plot(city[x])
name = ['Beijing', 'Chengdu', 'Guangzhou', 'Shanghai', 'Shenyang']
x_tick = []
for x in range (1, 32, 5):
    x_tick.append(x)
for x in range(0, len(ax)):
    ax[x].set_xticks(x_tick)
    ax[x].set_xticklabels(x_tick, rotation = 30, fontsize = 'small')
    ax[x].set_title(name[x] + '-2014-July')
fig7.subplots_adjust(wspace = 0.27, hspace = 0.35)
fig7.show()
```

**为了方便对比，将各个城市2014年7月的数据再绘制在一张图上**

```python
plt.style.use('default')
fig8 = plt.figure()
ax = fig8.add_subplot(111)
for x in city:
    ax.plot(x)
ax.legend(name, loc = 'best')
ax.set_title('PM in several cities (2014-7)')
```

绘制图像

```python
# 使用'seaborn'风格的图像
plt.style.use('seaborn')
y_name = ['Petal.Width', 'Petal.Length', 'Sepal.width', 'Sepal.length']
x_name = y_name[::-1]
fig10 = plt.figure()
# 给一个起始占位符，使得之后可以直接取相应下标的标号
# ax[i][j]表示第i行第j个子图
ax = [0]
for i in range(1, 5):
    line = [0]
    for j in range(1, 5):
        line.append(fig10.add_subplot(4, 4, (i - 1)  * 4 + j))
    ax.append(line)
    
# 隐藏横纵轴坐标
for x in range(1, 5):
    for y in range(1,5):
        if x == 4 and y ==1:
            pass
        elif x == 4:
            ax[x][y].set_yticks([])
        elif y == 1:
            ax[x][y].set_xticks([])
        else:
            ax[x][y].set_xticks([])
            ax[x][y].set_yticks([])
# 调整子图之间的间距
fig10.subplots_adjust(wspace = 0.05, hspace = 0.05)

# 绘制散点图
colors = ['blue', 'orange', 'green']
Species = iris.Species.unique()
for x in range(1, 5):
    for y in range(1, 5):        
        for i in range(len(Species)):
            ax[x][y].scatter(iris.loc[iris.Species == Species[i], y_name[y - 1]], iris.loc[iris.Species == Species[i], x_name[x - 1]], s = 20, c = colors[i], label = Species[i], alpha = 0.5)
        ax[x][y].grid(True, linestyle = '--', alpha = 0.8)

# 对角线的图中添加说明
ax[1][4].text(4.2, 7, x_name[0], fontsize = 'small')
ax[2][3].text(1.9, 3.8, x_name[1], fontsize = 'small')
ax[3][2].text(1.1, 5.6, x_name[2], fontsize = 'small')
ax[4][1].text(0.2, 2.1, x_name[3], fontsize = 'small')
```
