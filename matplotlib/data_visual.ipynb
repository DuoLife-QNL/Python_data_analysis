### 创建数据并查看其分布情况（尝试创建了正态分布）

```python
# 创建一个满足正态分布的数据，均值为700，标准差为100，规模10000个
samples = np.random.normal(loc = 500, scale = 100, size = 10000)
# 定义函数，完成对某个list中的数值进行统计，并返回一个dict
def count_elements(sample):
    value_count = {}
    for x in sample:
        value_count[int(x)] = value_count.get(int(x), 0) + 1
    return value_count
# 绘制柱状图
counted = count_elements(samples)
fig1 = plt.figure()
ax1 = fig1.add_subplot(111)
ax1.axis([0, 1000, 0, 60])
ax1.bar(counted.keys(), counted.values())
```

# 作业6

导入数据

```python
Beijing = pd.read_csv('..\data\BeijingPM20100101_20151231.csv')
Chengdu = pd.read_csv('..\data\ChengduPM20100101_20151231.csv')
Guangzhou = pd.read_csv('..\data\GuangzhouPM20100101_20151231.csv')
Shanghai = pd.read_csv('..\data\ShanghaiPM20100101_20151231.csv')
Shenyang = pd.read_csv('..\data\ShenyangPM20100101_20151231.csv')
haze = [Beijing, Guangzhou, Shenyang, Shanghai, Chengdu]
```

```python
# 数据预处理，丢弃空值数据，并计算平均值
# 由于几个城市的PM监测点数目不同，故需要传递描述PM指数的起始列和结束列
def data_pre(city, PM_col_start, PM_col_end):
    s_col = city.columns[1:4].append(city.columns[PM_col_start : PM_col_end + 1])
    PM_col = city.columns[PM_col_start : PM_col_end + 1]
    city = city.loc[:, s_col]
    city_cln = city.dropna(axis = 0, how = 'all', subset =PM_col).copy()
    city = city_cln
    city.loc[:, 'PM_AVG'] = np.nan
    for i in city.index:
        cnt_not_null = 0
        PM_sum = 0
        for j in city.columns[3 : PM_col_end - 2]:
            if not np.isnan(city.loc[i, j]):
                cnt_not_null += 1
                PM_sum += city.loc[i, j]
        city.loc[i, 'PM_AVG'] = PM_sum / cnt_not_null
    return city
```

```python
# 先处理北京（有四列数据和PM相关）
haze[0] = data_pre(haze[0], 6, 9)
# 再处理其它的城市（只有三列和PM有关）
for x in range(1, 5):
    haze[x] = data_pre(haze[x], 6, 8)
```

```python
# 按照天聚集
grouped = []
for x in range(0, 5):
    grouped.append(haze[x]['PM_AVG'].groupby([haze[x]['year'], haze[x]['month'], haze[x]['day']]).mean())
```

```python
df = []
for x in range(0, 5):
    df.append(pd.DataFrame(grouped[x]))
    for i in df[x].index:
        df[x].loc[i, 'date'] = str(i[0]) + '-' + str(i[1]) + '-' + str(i[2])
```

**将全部数据绘制到图上**

```python
fig6 = plt.figure()
bj = fig6.add_subplot(231)
cd = fig6.add_subplot(232)
gz = fig6.add_subplot(233)
sh = fig6.add_subplot(234)
sy = fig6.add_subplot(235)
bj.plot(df[0].date, df[0].PM_AVG)
cd.plot(df[1].date, df[1].PM_AVG)
gz.plot(df[2].date, df[2].PM_AVG)
sh.plot(df[3].date, df[3].PM_AVG)
sy.plot(df[4].date, df[4].PM_AVG)
bj.set_title('Beijing')
cd.set_title('Chengdu')
gz.set_title('Guangzhou')
sh.set_title('Shanghai')
sy.set_title('Shenyang')
```
